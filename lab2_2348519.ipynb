```python
# Calculating the Covariance Matrix

# We create a new dataset 'data' containing only the 'age' and 'Rating' columns as a NumPy array.
data = df[['age', 'Rating']].to_numpy()

# Compute the covariance matrix for the 'data' array using np.cov().
# The 'rowvar=False' argument indicates that each column represents a variable.
covariance_matrix = np.cov(data, rowvar=False)

# Printing the covariance matrix to analyze the relationships between variables.
print("Covariance Matrix:")
print(covariance_matrix)
```
